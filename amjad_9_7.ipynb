```python
import numpy as np

# Implement cosine similarity using the dot product
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    similarity = dot_product / (norm_v1 * norm_v2)
    return similarity

# Example vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

# Calculate cosine similarity
similarity = cosine_similarity(vector_a, vector_b)
print(f"Cosine similarity: {similarity}")
```

```python
# Install the sentence-transformers library if not installed
!pip install sentence-transformers

import numpy as np
from sentence_transformers import SentenceTransformer

# Load a more advanced pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

def get_sentence_embedding(sentence):
    return model.encode(sentence)

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def dot_product_similarity(v1, v2):
    return np.dot(v1, v2)

# Function to compare sentences
def compare_sentences(sentence1, sentence2):
    embedding1 = get_sentence_embedding(sentence1)
    embedding2 = get_sentence_embedding(sentence2)

    cosine_sim = cosine_similarity(embedding1, embedding2)
    dot_sim = dot_product_similarity(embedding1, embedding2)

    print(f"Sentence 1: '{sentence1}'")
    print(f"Sentence 2: '{sentence2}'")
    print(f"Cosine similarity: {cosine_sim:.4f}")
    print(f"Dot product similarity: {dot_sim:.4f}")
    print()

# changing .....>>>>>>>>>>>>>>>
compare_sentences("Books are my favorite companions", "Novels are my preferred companions")
compare_sentences("We studied hard for the exam", "For the exam, we studied hard")
compare_sentences("I really enjoy Chinese restaurants", "I love dining at Italian eateries.")
compare_sentences(" المشي في الهواء الطلق يساعد على تهدئة الأعصاب", " الجري في الصباح يجعلني أشعر بالنشاط والحيوية")

# Task for users:
print("Now it's your turn! Try comparing different sentences and have fun exploring similarities!")
print("1. Compare sentences with similar meanings but different words.")
print("2. Try sentences with the same words in different orders.")
print("3. Experiment with sentences from different domains or languages.")

# Reflection questions:
print("\nReflection questions:")
print('''1. How do cosine similarity and dot product similarity yield identical results in the given examples? What does this reveal about their mathematical properties?
In the examples provided, cosine similarity and dot product similarity yield identical results because the sentence embeddings are likely normalized vectors. When vectors are normalized
 (i.e., their magnitudes are normalized to 1), cosine similarity simplifies to the dot product of the vectors. This reveals that cosine similarity is a normalized version of dot product
 similarity, emphasizing the direction (cosine of the angle) rather than just the magnitude of vectors. This property makes cosine similarity particularly effective in measuring semantic
 similarity independent of the vector magnitudes.''')

print('\n')
print('''2. What types of sentences seem to have high similarity? Low similarity?
High Similarity:
Sentences with similar meanings or overlapping vocabulary tend to have high similarity. Examples like "Books are my favorite companions" and "Novels are my preferred companions" demonstrate
 high similarity due to their semantic equivalence despite different word choices.
Sentences with similar structure and order of words also exhibit high similarity, as seen in "We studied hard for the exam" and "For the exam, we studied hard".

Low Similarity:
Sentences that discuss unrelated topics or express contrasting sentiments typically have low similarity scores. For instance, "I really enjoy Chinese restaurants" and "I love dining at Italian
 eateries" show lower similarity due to the divergence in their discussed cuisines.
Sentences with different structures or word orders, even if conveying similar meanings, might also show lower similarity if the embedding model does not capture semantic equivalence well across
different syntactic forms.


''')

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate correlated random points
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix with correlation
points = np.random.multivariate_normal(mean, cov, 100)
X = points[:, 0].T
y = points[:, 1].T

# Plot the data points
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Correlated Random Points')
plt.show()

# Create the design matrix [X 1]
X_design = np.vstack((X, np.ones_like(X))).T

# Calculate the pseudo-inverse
X_pseudo_inv = np.linalg.pinv(X_design)

# Calculate the coefficients
coefficients = X_pseudo_inv.dot(y)

# Print the coefficients
print(f"Estimated coefficients: {coefficients.flatten()}")

# Plot the results
plt.scatter(X, y, label='Data points')
plt.plot(X, X_design.dot(coefficients), color='red', label='Linear model')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression using Pseudo-Inverse')
plt.show()
```
